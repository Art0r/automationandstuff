```python
top_mal = pd.read_html('https://myanimelist.net/topanime.php')
```

```python
top_anime = top_mal[0][[0, 1, 2]]
#top_anime[1]#.apply(lambda t: t)
```

```python
from typing import List
def get_TV(t: str) -> List[str]:
    if t != 'Title':
        ini_platform = None
        end_platform = t.find('s)')
        initial_month, end_month = None, None
        platform_type = ['OVA', 'TV', 'Movie']
        months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                  'Jul', 'Jun', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
        for i in platform_type:
            ini_platform = t.find(i)
            if ini_platform != -1:
                break
        for i in months:
            initial_month = t.find(i)
            if initial_month != -1:
                break
        for i in months:
            end_month = t.find(i)
            if end_month != -1 and end_month != initial_month:
                break
        title = t[:ini_platform:]
        platform = t[ini_platform:end_platform+2:]
        t = t.replace(title, '').strip()
        t = t.replace(platform, '').strip()
        release = t[:20:]
        t = t.replace(release, '').strip()
        members = t[:len(t):]
        return [title.strip(), platform.strip(), release.strip(), members.strip()]
    
#get_TV('Odd Taxi  TV (13 eps) Apr 2021 - Jun 2021 314,101 members')
get_TV('Koe no Katachi Movie (1 eps) Sep 2016 - Sep 2016 1,981,626 members')
```

```python
x=pd.DataFrame(columns=["RANK", "TITLE", "PLATFORM", "RELEASE", "MEMBERS", "SCORE"])
for index in top_anime.dropna().index:
    if index == 0: 
        continue
    x.loc[index, 'RANK'] = top_anime.loc[index, 0]
    x.loc[index, 'TITLE'] = get_TV(top_anime.loc[index, 1])[0]
    x.loc[index, 'PLATFORM'] = get_TV(top_anime.loc[index, 1])[1]
    x.loc[index, 'RELEASE'] = get_TV(top_anime.loc[index, 1])[2]
    x.loc[index, 'MEMBERS'] = get_TV(top_anime.loc[index, 1])[3]
    x.loc[index, 'SCORE'] = top_anime.loc[index, 2]
    
x
```
